```python
#메모리 초과
import sys
N = int(sys.stdin.readline())
nodes = []
for i in range(N):
    nodes.append(tuple(map(int,sys.stdin.readline().split())))
edges = []
for i in range(N):
    for j in range(N):
        if i > j :
            continue
        if i != j:
            edges.append((i,j,min(abs(nodes[i][0]-nodes[j][0]),abs(nodes[i][1]-nodes[j][1]),abs(nodes[i][2]-nodes[j][2]))))
edges.sort(key=lambda x: x[2])
parents = [ i for i in range(N)]
def find(a):
    if parents[a] == a :
        return a
    else:
        parents[a] = find(parents[a])
        return parents[a]
def union( a, b):
    rootA = find(a)
    rootB = find(b)
    if rootA != rootB:
        if rootA < rootB:
            parents[rootB] = rootA
        else:
            parents[rootA] = rootB
        return True
    else:
        return False
count = 0
result = 0
index= 0
while count!=N-1:
    a,b,c = edges[index]
    if union(a,b) :
        result+=c
        count+=1
    index+=1
print(result)
```

```python
import sys
C = int(input())
for c in range(C):
    n = int(sys.stdin.readline())
    temp = list(map(int,sys.stdin.readline().split()))
    for i,v in enumerate(temp):
        temp[i] = (v,i)
    temp.sort()
    graph = [[] for i in range(n)]
    for index in range(len(temp)-1):
        graph[temp[index][1]].append(temp[index+1][1])
    index = []
    for i in range(n):
        index.append([1,i])
    index[temp[0][1]][0] = 0
    #여기서 그래프는 랭킹이 높은애를 가리킨다.
    m = int(sys.stdin.readline())
    for i in range(m):
        a,b = map(int,sys.stdin.readline().split())
        if b in graph[a]:
            graph[a].remove(b)
            graph[b].append(a)
            index[a] = max(index[a]-1,0)
            index[b] +=1
        else:
            graph[b].remove(a)
            graph[a].append(b)
            index[b] = max(index[b]-1,0)
            index[a] +=1
    index.sort()
```
