```python
#금광
#대충 뒤에서부터 dp
import sys
T = int(input())
results=[]
for i in range(T):
    n,m = map(int,sys.stdin.readline().split())
    nums = list(map(int,sys.stdin.readline().split()))
    MAP = [[0 for x in range(m)] for k in range(n)]
    for index,val in enumerate(nums):
        MAP[index//m][index%m] = val
    result = 0
    for col in range(1,m):
        for row in range(0,n):
            drs = [-1,0,1]
            temp = MAP[row][col]
            for dr in drs:
                if 0<=dr+row<n:
                    temp = max(MAP[row][col]+MAP[dr+row][col-1],temp)
            MAP[row][col]=temp
    for w in range(n):
        result = max(MAP[w][-1],result)
    results.append(result)
for i in results:
    print(i)
        
```

```python
#퇴사 그래프로 구하기 
import sys
def dfs(start,p):
    result = p
    for i in graph[start]:
        result = max(dfs(i,p+val[i]),result)
    return result
N = int(input())
val = [ 0 for i in range(N+1)]
graph = [[] for i in range(N+1)]
for i in range(1,N+1):
    t,p = map(int,sys.stdin.readline().split())
    if i+t < N+1:
        graph[i+t].append(i)
        val[i]=p
result = 0
indice = [i for i in range(1,N+1)]
for i in indice[::-1]:
    if len(graph[i]) !=0:
        result = max(result,dfs(i,val[i]))
print(result)
```

```python
import sys
n = int(input())
MAP = []
def padding(n,index,data):
    add = [0 for i in range(n-index+1)]
    return data+add
for i in range(n):
    temp = list(map(int,sys.stdin.readline().split()))
    MAP.append(padding(n,i,temp))
for i in range(1,n):
    for j in range(i+1):
        temp = MAP[i][j]
        for dx in [-1,0]:
            if j+dx >=0 :
                temp = max(temp,MAP[i-1][j+dx]+MAP[i][j])
        MAP[i][j]=temp
result = 0
for i in range(n):
    result = max(MAP[-1][i],result)
print(result)
```

```python
#가장 긴 증가하는 부분수열 구하기
import sys
n = int(input())
data = list(map(int,sys.stdin.readline().split()))
data = data[::-1]

dp = [1] * n
for i in range(1,n):
    for j in range(0,i):
        if data[j] < data[i]:
            dp[i] = max(dp[i], dp[j]+1)

print(n-max(dp))
```
