```python
def distance(idx1,idx2,weak,n): 
    # length, is_clock? 
    clock = weak[idx2] - weak[idx1]
    counter = weak[idx2] + n - weak[idx1]
    if clock > counter : 
        return clock, True
    elif clock < counter:
        return counter, False
    else:
        if len(weak[idx1:idx2+1]) > len(weak[idx2:]) + len(weak[:idx1+1]) :
            return clock, True
        else:
            return counter , False
```

```python
from itertools import permutations
def solution(n,weak,dist):
    length = len(weak)

    for i in range(length):
        weak.append(weak[i]+n)
    
    answer = len(dist) + 1

    for start in range(length):
        for friends in list(permutations(dist,len(dist))):
            count = 1
            position = weak[start] + friends[count-1]

            for index in range(start,start+ length):
                if position < weak[index]:
                    if count>len(dist) :
                        break
                    position = weak[index] + friends[count-1]
            answer = min(answer, count)
    if answer > len(dist):
        return -1
    else:
        return answer
        
```

```python
import sys
from collections import Counter
N,L,R = map(int,sys.stdin.readline().split())
A = []
for i in range(N):
    A.append(list(map(int,sys.stdin.readline().split())))
roots = [ i for i in range(N*N)]
def coordinate(i):
    global N
    return i//N,i%N #row, column
count = 0
dx=[0,0,-1,1]
dy=[1,-1,0,0]

while True:
    union = [[] for i in range(N*N)]
    roots = [i for i in range(N * N)]
    for i in range(N*N):
        union[i].append(i)
    for i in range(N*N):
        r,c = coordinate(i)
        for ddx,ddy in zip(dx,dy):
            acr = r+ddx
            acc = c+ddy
            if 0<=acr<N and 0<=acc<N :
                diff = A[r][c] -A[acr][acc]
                if L<=abs(diff)<=R:
                    if roots[acr*N+acc] != acr*N+acc:
                        roots[i] = roots[acr*N+acc]
                        union[roots[acr*N+acc]] +=union[roots[i]]
                    else:
                        roots[acr*N+acc] = roots[i]
                        union[roots[i]] += union[acr * N + acc]
    groups = Counter(roots)
    del roots
    end = True
    for key,value in groups.items():
        if value > 1:
            result=0
            for i in set(union[key]):
                r,c=coordinate(i)
                result+=A[r][c]
            for i in set(union[key]):
                r,c = coordinate(i)
                A[r][c] = result//value
            end = False
    if end :
        break
    else:
        count+=1
print(count)
```
